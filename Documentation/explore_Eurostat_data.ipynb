# Bilan energétique complet - Eurostat nrg_bal_c.tsv

## Exploration

```python
df = pd.read_csv('nrg_bal_c.tsv', sep='\t')
df
```

```python
df.columns.values
```

## Split de la première colonne

```python
df['nrg_bal,siec,unit,geo\\time']
```

```python
text_columns = df[['nrg_bal,siec,unit,geo\\time']]
text_columns
```

```python
text_columns[['nrg_bal', 'siec', 'unit', 'geo\\time']] = text_columns['nrg_bal,siec,unit,geo\\time'].str.split(",",expand=True,)
text_columns
```

```python
new_columns = text_columns[['nrg_bal', 'siec', 'unit', 'geo\\time']]
new_columns
```

## Ajout des nouvelles colonnes au dataframe

```python
df1 = pd.concat([df, new_columns], axis=1)
df1
```

```python
df1 = df1.drop(columns=['nrg_bal,siec,unit,geo\\time'])
df1
```

## Sélection de l'unité en Kilo Tonne équivalent pétrole

```python
df1 = df1.rename(columns={'geo\\time':'country'})
df2 = df1[df1['unit']=='KTOE']
df2
```

```python
df2.columns
```

## Création dataframe final avant prétraitement des données

### Création de plusieurs dataframe en fonction des années et ajout d'une colonne année

```python
df_bilan_energ = []
for i in range(1990,2019):
    df_i = df2[[f'{i} ','nrg_bal', 'siec', 'unit','country']]
    df_i.rename(columns={f'{i} ':'quantite'}, inplace=True)
    df_i['annee'] = i
    df_bilan_energ.append(df_i)
df_bilan_energ
```

```python
len(df_bilan_energ)
```

### Création du dataframe final selon le modèle du schema DB

```python
df_bilan_final =pd.concat([df_bilan_energ[0],df_bilan_energ[1],df_bilan_energ[2],df_bilan_energ[3],
                           df_bilan_energ[4],df_bilan_energ[5],df_bilan_energ[6],df_bilan_energ[7],
                          df_bilan_energ[8],df_bilan_energ[9],df_bilan_energ[10],df_bilan_energ[11],
                          df_bilan_energ[12],df_bilan_energ[13],df_bilan_energ[14],df_bilan_energ[15],
                          df_bilan_energ[16],df_bilan_energ[17],df_bilan_energ[18],df_bilan_energ[19],
                          df_bilan_energ[20],df_bilan_energ[21],df_bilan_energ[22],df_bilan_energ[23],
                          df_bilan_energ[24],df_bilan_energ[25],df_bilan_energ[26],df_bilan_energ[27],
                          df_bilan_energ[28]])
df_bilan_final
```

# Table Dechets - ten00106+ESTAT.tsv

Structure du fichier selon le même modèle que le fichier sur le bilan énergétique:
- Problématique de création des colonnes correspondantes à la première colonne du fichier
- Création de sous-dataframe selon les années
- Concaténation des sous-dataframe selon le modèle du schema DB

```python
df = pd.read_csv('ten00106.tsv', sep='\t')
df
```

## Split de la première colonne

```python
first_col = df[['freq,unit,hazard,nace_r2,waste,geo\TIME_PERIOD']]
first_col
```

```python
first_col[['freq','unit','hazard','nace_r2','waste','geo\TIME_PERIOD']] = first_col['freq,unit,hazard,nace_r2,waste,geo\TIME_PERIOD'].str.split(",",expand=True,)
first_col
```

```python
newcol = first_col[['freq','unit','hazard','nace_r2','waste','geo\TIME_PERIOD']]
newcol = newcol.drop(columns=['freq','unit','hazard','waste'])
newcol.rename(columns={'geo\TIME_PERIOD':'country'}, inplace=True)
newcol
```

```python
df_waste = pd.concat([df, newcol], axis=1)
df_waste = df_waste.drop(columns=['freq,unit,hazard,nace_r2,waste,geo\TIME_PERIOD'])
df_waste
```

```python
df_waste.columns
```

```python
# test de création d'un sous dataframe
df_2004 = df_waste[['2004 ','nace_r2','country']]
df_2004.rename(columns={'2004 ':'quantite(Tonne)'}, inplace=True)
df_2004['annee'] = 2004
df_2004
```

## Création des sous dataframe

```python
df_global = []
for i in range(2004,2020,2):
    df_i = df_waste[[f'{i} ','nace_r2','country']]
    df_i.rename(columns={f'{i} ':'quantite(Tonne)'}, inplace=True)
    df_i['annee'] = i
    df_global.append(df_i)
df_global[0]
```

```python
len(df_global)
```

## Concatenation des sous dataframe puis création de la table

```python
df_waste_final =pd.concat([df_global[0],df_global[1],df_global[2],df_global[3],df_global[4],df_global[5],df_global[6],df_global[7]])
df_waste_final.rename(columns={'nace_r2':'code_act_eco','country':'pays'}, inplace=True)
```

```python
df_waste_final.to_csv('table_dechet.tsv', index=False, header=['quantite(Tonne)','code_act_eco','pays','annee'], sep='\t')
```
